```python
import pandas as pd
import numpy as np
data = pd.read_csv('listings_berlin.csv')
data.head()

#Task3
data['price'] = data['price'].str.replace('$','')
data['price'] = data['price'].str.replace(',','')
data['price'] = data['price'].astype(float)

#Task 4
data['bathrooms_text'].value_counts()
def clean_baths(bathroom):
    if pd.isna(bathroom):
        return None
    
    bathroom = str(bathroom).lower()
    if 'half' in bathroom:
        return 0.5
    else:
        numpart = bathroom.split(' ')[0]
        return float(numpart)
    '''Firstly we check if the value is NaN, if it is we return None. 
    Then we make value lowercase. If the string contains 'half', we return 0.5. 
    Otherwise, we split the string by space and take the first part, which should be the number of bathrooms, 
    and convert it to a float before returning it. Afterwards we apply this function for a column baths_text'''
data['bathrooms_text'] = data['bathrooms_text'].apply(clean_baths)

#Task 8
def trimmed_mean(x, pct):

    x = sorted(x)
    n = int(len(x) * (pct/100))
    for i in range(n):
        del x[0]
        del x[-1]
    
    return sum(x) / len(x)
cleaned_prices = data['price'].dropna().to_numpy()
trim_m = trimmed_mean(cleaned_prices, 5)
arith_m = cleaned_prices.mean()
compare = trim_m / arith_m
print(f'Trimmed mean: {trim_m} Aritmetic mean: {arith_m} Comparison: {compare}')
'''We define a function trimmed_mean with 2 parameterrs: x - list of numbers, pct - percentage
we sort the list x, calculate n - number of elements to remove from each side. 
We use a loop to remove n smallest and n largest elements from the list and return the mean of a new list. 
Then we apply this function to the cleaned_price columns that does not contain NaN values. 
Finally we calculate the arithmetic mean of the same column and compare it to the trimmed mean.
'''
```
